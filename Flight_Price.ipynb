```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df['Date']= df['Date_of_Journey'].str.split('/').str[0]
df['Month']= df['Date_of_Journey'].str.split('/').str[1]
df['Year']= df['Date_of_Journey'].str.split('/').str[2]
```

```python
df['Date_of_Journey'].str.split('/').str[0]
```

```python
df.info()
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
##drop the Date_of_Journey column
df.drop('Date_of_Journey', axis=1, inplace=True)
df.info()
```

```python
df.head(2)
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_hour']= df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min']= df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
df.drop('Arrival_Time',axis=1, inplace= True)
```

```python
df.info()
```

```python
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
```

```python
df.head(2)
```

```python
df['Dep_hour']= df['Dep_Time'].str.split(':').str[0]
df['Dep_min']= df['Dep_Time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
df.drop('Dep_Time', axis=1, inplace=True)
```

```python
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
df.info()
```

```python
df['Duration_hour']= df['Duration'].str.split('h').str[0]
df['Duration_min']= df['Duration'].str.split(' ').str[1]
df.head(2)
```

```python
df['Duration_min'] = df['Duration_min'].str.split('m').str[0]
df['Duration_min'] = df['Duration_min'].fillna('0').astype(int)
```

```python
# Remove any non-numeric characters and replace empty strings with 0
df['Duration_hour'] = df['Duration_hour'].replace('', '0')
df['Duration_hour'] = df['Duration_hour'].str.extract('(\d+)').fillna(0).astype(int)
df.head(2)
```
